```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = [10, 12, 15, 18, 20, 22, 25, 30, 35, 40]
df = pd.DataFrame(data, columns=["values"])
print("Dataset:")
print(df)
print("\nSummary Statistics:")
print(df.describe())
print("\nVariance:", np.var(data))
print("Standard Deviation:", np.std(data))
plt.hist(data)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Data Distribution")
plt.show()
mean_value = np.mean(data)
median_value = np.median(data)
print("\nInference")
print("Mean:", mean_value)
print("Median:", median_value)
if mean_value > median_value:
    print("Data is slightly right-skewed")
elif mean_value < median_value:
    print("Data is slightly lef")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
url = "https://raw.githubusercontent.com/dataprofessor/data/master/BostonHousing.csv"
df = pd.read_csv(url)
print("Dataset (first 10 rows):")
print(df.head(10))
data = df["medv"]
print("In summary statistics:")
print(data.describe())

print("In Variance:", np.var(data))
print("Standard deviation:", np.std(data))

plt.hist(data, bins=10, color='skyblue', edgecolor='black')
plt.axvline(np.mean(data), color='red',
            linestyle='dashed',
            linewidth=2,
            label=f"Mean: {np.mean(data):.2f}")
plt.axvline(np.median(data), color='green',
            linestyle='dashed',
            linewidth=2,
            label=f"Median: {np.median(data):.2f}")
plt.xlabel("Median Home Value (medv)")
plt.ylabel("frequency")
plt.title("Distribution of Median Home Value")
plt.legend()
plt.show()
mean_value = np.mean(data)
median_value = np.median(data)
print("In Inter_Inference:")
print("Mean value:", mean_value)
print("Median value:", median_value)
if mean_value > median_value:
    print("Data is slightly right_skewed")
elif mean_value < median_value:
    print("Data is slightly left-skewed")
else:
    print("Data is symmetric")
```
